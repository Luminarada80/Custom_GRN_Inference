```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn

run_output_dir = "/gpfs/Labs/Uzun/SCRIPTS/PROJECTS/2024.SINGLE_CELL_GRN_INFERENCE.MOELLER/output/transformer_testing_output/training_stats"


run_loss_dataframes = {}
for file in os.listdir(run_output_dir):
    training_file = os.path.join(run_output_dir, file)
    file_no_ext = os.path.splitext(file)[0]
    
    if not file in run_loss_dataframes:
        run_loss_dataframes[file_no_ext] = pd.read_csv(training_file, sep=",", header=0, index_col=0)
    
for run_name, run_df in run_loss_dataframes.items():
    print(run_name)
    
```

```python
chr19_run_df = run_loss_dataframes["chr19_training_loss_window_1kb_500_cells"]
chr19_run_df.head()
```

```python
fig = plt.figure(figsize=(7, 6))
plt.plot(chr19_run_df.index, chr19_run_df["val_loss"], linewidth=2, label="Validation Loss")
plt.plot(chr19_run_df.index, chr19_run_df["train_loss"], linewidth=2, label="Training Loss")

plt.title("mESC chr19 training/validation loss \n1kb windows 500 cells", fontsize=16)
plt.ylim((0,1))
plt.yticks(fontsize=14)
plt.xticks(fontsize=14)
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("Loss", fontsize=14)
plt.legend(fontsize=14)
plt.tight_layout()
fig.show()
```
